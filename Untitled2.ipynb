```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#LSP dataset
labels = ['Head', 'Shoulder', 'Elbow', 'Wrist', 'Hip' , 'Knee', 'Ankle', 'Mean']
men_means = [95.35, 87.47, 81.37, 77.95, 84.54, 85.17, 84.84, 85.29]#exp5
women_means = [98.50, 94.00, 89.80, 87.50, 93.90, 94.10, 93.00, 93.10]#state-of-the-art
child_means = [92.14, 84.86, 79.21, 77.14, 81.79, 81.81, 80.46, 82.40] #exp2

x = np.arange(len(labels))  # the label locations
width = 0.20  # the width of the bars

fig, ax = plt.subplots(figsize = (60 ,30))
rects1 = ax.bar(x - width/3, child_means, width, label='EXP2')
rects2 = ax.bar(x + 2*width/3, men_means, width, label='EXP6')
rects3 = ax.bar(x + 5* width/3, women_means, width, label='STATE-OF-THE-ART')



# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Test accuracy(%)', fontsize= 50)
ax.set_title('Comparison the test accuracy of co-ordinate joints with different model', fontsize=50)
ax.set_xticks(x)
ax.set_xticklabels(labels , fontsize= 50)
ax.legend(fontsize= 30)
ax.tick_params(labelsize=50)


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 3, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize= 35)


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)



fig.tight_layout()

plt.show()
```
