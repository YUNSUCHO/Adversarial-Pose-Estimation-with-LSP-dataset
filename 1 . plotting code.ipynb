```python
from os.path import join
import argparse

from glob import glob
import cv2
from scipy.io import loadmat
import torch
from torch.utils.data import Dataset
from matplotlib import pyplot as plt
import numpy as np
```

```python
path_s  ='/home/ycho/Adversarial-Pose-Estimation.lsp.V.4/lspet_dataset' 
path ='/home/ycho/Adversarial-Pose-Estimation.lsp.V.4/lsp_dataset/lsp_dataset'



annot = np.concatenate((loadmat(join(path, 'joints.mat'))['joints'].transpose(1 , 0 , 2), 
                                            loadmat(join(path_s, 'joints.mat'))['joints']) , axis =2)
```

```python
arr =[]
for idx in range(0 , 12000):
    for part in range(len(annot)):
        
        if part == 0  and   annot[0][0][idx] !=0:
            arr.append("Right ankle")
        elif part == 1  and annot[1][0][idx] !=0:
            arr.append("Right knee")
        elif part == 2  and annot[2][0][idx] !=0:
            arr.append("Right hip")
        elif part == 3  and annot[3][0][idx] !=0:
            arr.append("Left hip")
        elif part == 4  and annot[4][0][idx] !=0:
            arr.append("Left knee")
        elif part == 5  and annot[5][0][idx] !=0:
            arr.append("Left ankle")
        elif part == 6  and annot[6][0][idx] !=0:
            arr.append("Right wrist")
        elif part == 7  and annot[7][0][idx] !=0:
            arr.append("Right elbow")
        elif part == 8  and annot[8][0][idx] !=0:
            arr.append("Right shoulder")
        elif part == 9  and annot[9][0][idx] !=0:
            arr.append("Left shoulder")
        elif part == 10  and annot[10][0][idx] !=0:
            arr.append("Left elbow")
        elif part == 11  and annot[11][0][idx] !=0:
            arr.append("Left wrist")
        elif part == 12  and annot[12][0][idx] !=0:
            arr.append("Neck")
        elif part == 13  and annot[13][0][idx] !=0:
            arr.append("Head top")
    
print(len(arr))
    
    
```

```python
import seaborn as sns
sns.set_style("darkgrid")
fig_dims = (20, 15)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot( y = arr)
plt.title('TOATAL NUMBER OF KEYPOINTS IN  THE LSP DATSET FOR EACH OF THE 14 LANDMARK ',
         fontsize =24 
         )
plt.xlabel ('COUNTS',fontsize =24)
plt.ylabel ('NAME OF THE LANDMARK',fontsize =24)
for p in ax.patches:
    percentage ='{:,.0f}'.format(p.get_width())
    width, height =p.get_width(),p.get_height()
    x=p.get_x()+width+0.02
    y=p.get_y()+height/2
    ax.annotate(percentage,(x,y), fontsize =24)
```
